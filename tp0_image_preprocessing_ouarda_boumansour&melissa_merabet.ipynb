```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import random
import numpy as np
import kagglehub

dataset_path = kagglehub.dataset_download("msarmi9/food101tiny")

image_files = []
for root, _, files in os.walk(dataset_path):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_files.append(os.path.join(root, file))

if image_files:
    img = Image.open(image_files[0])
    print(f"Image dimensions: {img.size}")

    # Display the first image
    plt.figure(figsize=(15, 5))
    plt.subplot(111)
    plt.imshow(img)
    plt.title("First image display")
    plt.axis('off')
    plt.show()

    num_images_to_display = 5
    selected_images = random.sample(image_files, min(num_images_to_display, len(image_files)))

    # Display the images
    plt.figure(figsize=(15, 5))
    for i, img_path in enumerate(selected_images):
        img = Image.open(img_path)
        plt.subplot(1, num_images_to_display, i + 1)
        plt.imshow(img)
        plt.title(os.path.basename(os.path.dirname(img_path)))
        plt.axis('off')
    plt.show()

    img_to_normalize = Image.open(image_files[0])

    # Convert the image to a NumPy array
    img_array = np.array(img_to_normalize)

    # Normalize the pixel values to the range [0, 1]
    normalized_img_array = img_array / 255.0

    print("Original image array shape:", img_array.shape)
    print("Normalized image array shape:", normalized_img_array.shape)
    print("Normalized image array (first few pixels):\n", normalized_img_array[:5, :5, :])

    # Calculate the mean and standard deviation of the normalized image array
    mean_pixels = np.mean(normalized_img_array)
    std_pixels = np.std(normalized_img_array)

    print(f"Mean of pixel values: {mean_pixels}")
    print(f"Standard deviation of pixel values: {std_pixels}")

else:
    print("No image files found in the dataset.")
```



mean is close to 0, the image is a bit dark.
a low standard deviation indicates that most pixel values are close to the mean (less contrast).
