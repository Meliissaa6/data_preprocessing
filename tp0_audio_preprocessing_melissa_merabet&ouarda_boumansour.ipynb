```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np

audio_path = '/content/1-69641-A.ogg'

# 1. Load the audio file
y, sr = librosa.load(audio_path)

# 2. Display the signal size and duration
signal_size = y.shape[0]
duration = signal_size / sr

print(f"Signal size: {signal_size} samples")
print(f"Duration: {duration:.2f} seconds")

# 3. Plot the waveform
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

# 4. Calculate and display the spectrogram
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.figure(figsize=(14, 5))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.title('Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.colorbar(format='%+2.0f dB')
plt.show()

# 5. Play the sound in Jupyter
print("Playing audio:")
ipd.Audio(y, rate=sr)
```
